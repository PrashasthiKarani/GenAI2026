```python
#Genai class4 Pandas

import pandas as pd

df = pd.DataFrame({

    'Name': ['A','B','C','D','E','F','G','H'],

    'Age': [23,45,31,35,62,18,29,40],

    'Gender': ['M','F','M','F','F','M','M','F'],

    'Salary': [50000,60000,55000,65000,70000,40000,52000,58000],

    'Department': ['HR','IT','IT','HR','Finance','IT','HR','Finance']

})
print("First 5 rows:")
print(df.head())
print("\nLast 5 rows:")
print(df.tail())
```

```python
# Correct way to assign Bonus based on Department column
df['Bonus'] = df.apply(lambda row: row['Salary'] * 0.2 if row['Department'] == 'HR' else row['Salary'] * 0.1, axis=1)
df.head()
```
