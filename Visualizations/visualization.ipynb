```python
# SIMPLE VISUALIZATIONS ON IRIS DATASET
# ====================================
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
# Load iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
print(df.head())
```

```python
#BAR chart

species_counts = df['species'].value_counts()
plt.figure(figsize=(8, 5))
plt.title('Count of Each Iris Species')
plt.bar(species_counts.index, species_counts.values, color=['blue', 'orange', 'green'])
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(species_counts.index, iris.target_names)
plt.show()
```

```python
#BAR chart

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Calculate mean of each feature per species
grouped = df.groupby('species').mean()

# Plot setup
x = np.arange(len(iris.target_names))  # [0, 1, 2] for 3 species
width = 0.2                             # width of each bar

fig, ax = plt.subplots(figsize=(10, 6))

# One bar group per feature
bars1 = ax.bar(x - 1.5*width, grouped['sepal length (cm)'], width, label='Sepal Length', color='steelblue')
bars2 = ax.bar(x - 0.5*width, grouped['sepal width (cm)'],  width, label='Sepal Width',  color='orange')
bars3 = ax.bar(x + 0.5*width, grouped['petal length (cm)'], width, label='Petal Length', color='green')
bars4 = ax.bar(x + 1.5*width, grouped['petal width (cm)'],  width, label='Petal Width',  color='red')

# Labels & formatting
ax.set_title('Average Feature Measurements per Iris Species', fontsize=14)
ax.set_xlabel('Species')
ax.set_ylabel('Average (cm)')
ax.set_xticks(x)
ax.set_xticklabels(iris.target_names)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
#pie chart 
species_counts.plot(kind='pie', autopct='%1.1f%%',startangle=90, colors=['steelblue', 'orange', 'grey'])
plt.title('Proportion of Each Iris Species')
plt.ylabel('')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
#LINE CHART
plt.plot(df['sepal length (cm)'], label='Sepal Length', color='steelblue')
plt.plot(df['sepal width (cm)'], label='Sepal Width', color='orange')
plt.plot(df['petal length (cm)'], label='Petal Length', color='green')
plt.plot(df['petal width (cm)'], label='Petal Width', color='red')
plt.title('Feature Measurements of Iris Dataset', fontsize=14)
plt.xlabel('Sample Index')
plt.ylabel('Measurement (cm)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# 4 BOXPLOT

# ====================================

sns.boxplot(x='species',
            y='sepal length (cm)',
            data=df)
plt.title("Boxplot")
plt.show()
```

```python
# 5 HEAT MAP
# ====================================
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Heatmap")
plt.show()
```
